```python
# Iloc method
X=data.iloc[:,:-1].values               #Input Var
Y=data.iloc[:,-1].values                # Target var
```

```python
X
```

```python
Y
```

```python
# Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# apply frist logistic regression
from sklearn.linear_model import LogisticRegression
lr= LogisticRegression()
import warnings
warnings.filterwarnings('ignore')
```

```python
lr.fit(X_train, Y_train)
```

```python
pred=lr.predict(X_test)
```

```python
pred
```

```python
Y_test
```

```python
#evaluatuion of ML model
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
print(confusion_matrix(Y_test, pred))
```

```python
print('test accurace score', accuracy_score(Y_test, pred))
```

```python
print('Train accurqace score', accuracy_score(Y_train, lr.predict(X_train)))
```

```python
print(classification_report(Y_test,pred))
```

```python
#prepare AUC and ROC Score
from sklearn.metrics import roc_auc_score, roc_curve
```

```python
prob= lr.predict_proba(X_test)
prob_pos=prob[:,1]
```

```python
auc_score=roc_auc_score(Y_test, prob_pos)
print('Auc score is', auc_score)
```

```python
fpr, tpr,thrs=roc_curve(Y_test, prob_pos)

plt.plot([0,1], [0,1], linestyle= '--')
plt.plot(fpr,tpr, marker='.')
plt.show()
```

```python
# save the model
import joblib
joblib.dump(lr, 'Log_Reg.pkl')
print('model saved')
```
